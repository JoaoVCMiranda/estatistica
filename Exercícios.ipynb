```python
# Lista 1 (Colunas da esquerda)
lista1 = [17, 16, 21, 14, 18, 24, 16, 14, 21, 23, 13, 18]

# Lista 2 (Colunas da direita)
lista2 = [18, 14, 19, 11, 23, 21, 10, 13, 19, 24, 15, 20]

m1 = media(lista1)
m2 = media(lista2)

n = len(lista1)

d_barra =  m1-m2
lista3 = [lista1[i]-lista2[i] for i in range(n)]

t_calculado = (d_barra-0) / ((variancia(lista3))**0.5/n)

print(n)
print("t calculado = ", t_calculado)
# sabendo que o t(12-1) para um grau de significância de 95% é 1.796
```

# Hipótese
$$
H_0: \Delta = 0 
$$
Portanto,
$$
t_{n-1} = \frac{\bar d - \Delta}{\frac{s_d}{\sqrt{n}}} = t_{n-1} = \frac{\bar d - 0}{\frac{s_d}{\sqrt{n}}} 
$$
